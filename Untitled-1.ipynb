```python
from PIL import Image
import requests
from io import BytesIO

# Correct URLs for raw images
original_image_url = 'https://raw.githubusercontent.com/aishuaishu10/aishwarya-pan-card/main/jnanavi.jpg'  # Pan card image
tampered_image_url = 'https://assets1.cleartax-cdn.com/s/img/20170526124335/Pan4.png'  # Tampered image

# Fetch images from URLs
original = Image.open(BytesIO(requests.get(original_image_url).content))
tampered = Image.open(BytesIO(requests.get(tampered_image_url).content))

# Image Formats
print(f"Original image format: {original.format}")
print(f"Tampered image format: {tampered.format}")

# Image sizes
print(f"Original image size: {original.size}")
print(f"Tampered image size: {tampered.size}")
```

```python
# Resize Images
img_size = (250, 160)
original = original.resize(img_size)
tampered = tampered.resize(img_size)
```

```python
# Save Images in jpg format
original.save('images/original.jpg')
tampered.convert('RGB').save('images/tampered.jpg')

# Load Images
original = Image.open('images/original.jpg')
tampered = Image.open('images/tampered.jpg')
```

```python
from skimage.io import imread
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

# Load the images using the correct relative paths
original = imread('images/original.jpg')
tampered = imread('images/tampered.jpg')

# Convert to grayscale (scale to [0, 255] and cast to uint8)
original_gray = (rgb2gray(original) * 255).astype('uint8')
tampered_gray = (rgb2gray(tampered) * 255).astype('uint8')

# Display the grayscale original image
plt.imshow(original_gray, cmap='gray')
plt.title("Original Image (Grayscale)")
plt.show()

# Display the grayscale tampered image
plt.imshow(tampered_gray, cmap='gray')
plt.title("Tampered Image (Grayscale)")
plt.show()
```

```python
from skimage.metrics import structural_similarity
import matplotlib.pyplot as plt

# Calculate SSIM (Structural Similarity Index)
score, diff_img = structural_similarity(original_gray, tampered_gray, full=True)
diff_img = (diff_img * 255).astype("uint8")

print(f"SSIM: {score}")

# Display the difference image
plt.imshow(diff_img, cmap='gray')
plt.title("Difference Image")
plt.axis('off')
plt.show()
```

```python
# Display difference image
Image.fromarray(diff_img)
```

```python
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
from skimage.metrics import structural_similarity
import matplotlib.pyplot as plt

# Load images
original = imread('images/original.jpg')
tampered = imread('images/tampered.jpg')

# Convert to grayscale
original_gray = (rgb2gray(original) * 255).astype('uint8')
tampered_gray = (rgb2gray(tampered) * 255).astype('uint8')

# Compute SSIM
score, diff_img = structural_similarity(original_gray, tampered_gray, full=True)
diff_img = (diff_img * 255).astype('uint8')

print(f"SSIM: {score}")

# Apply Otsu threshold
binary_img = diff_img <= threshold_otsu(diff_img)

# Show the binary image
plt.imshow(binary_img, cmap='gray')
plt.title("Binary Difference Image (Otsu Threshold)")
plt.axis('off')
plt.show()
```
